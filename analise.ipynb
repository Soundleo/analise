Nesse exempo, vamos analisar o Dataset da taxa de cancelamentos de panos dos usários da empresa farmaceutica **Ubrella Corporation**


Trazendo os dados do Dataset para o notebook

```python
#importa a biblioteca pandas
import pandas as pd 

#Lê aos dados do database
tabela = pd.read_csv('cancelamentos.csv')

#mostra o dataset tela
display(tabela)
```

```python
tabela = tabela.drop('CustomerID', axis=1)
display(tabela)
```

```python
display(tabela.info())
```

```python
tabela = tabela.dropna()
display(tabela.info())
```

Vamos contar quantos usuários cancelarem e quantos não canceleram 

```python
display(tabela['cancelou'].value_counts())
```

Traduzindo para porcentagem 

```python
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```

Proxima analise de dados será para verificar a duração dos contratos

```python
display(tabela['duracao_contrato'].value_counts())
display(tabela['duracao_contrato'].value_counts(normalize=True).map('{:.2%}'.format))
```

```python
display(tabela.groupby('duracao_contrato').mean(numeric_only=True))
```

Vamos eliminar is dados dos contratos mensais. 

```python
tabela = tabela[tabela['duracao_contrato'] != 'Monthly']
display(tabela)
display(tabela['cancelou'].value_counts())
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```

Vamos verificar os tipos de assinaturas
---

```python
display(tabela['assinatura'].value_counts(normalize=True).map('{:.2%}'.format))
display(tabela.groupby('assinatura').mean(numeric_only=True))
```

Vamos passar a nalisar os dados das assinaturas olhado os gráficos de cada coluna   

```python
import plotly.express as px

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou", width=600)
    grafico.show()
```

Localizado anomalias: ligações para o call center e dias de atraso. Os clientes com mais de 20 dias de atraso  canceleram as assinaturas, enquanto que os clientes que fizeram mais de 5 ligações no call center, todos canceleram as suas assinaturas. 

```python
tabela = tabela[tabela['ligacoes_callcenter'] < 5]
tabela = tabela[tabela['dias_atraso']<=20]
display(tabela)
display(tabela['cancelou'].value_counts())
display(tabela['cancelou'].value_counts(normalize=True).map('{:.2%}'.format))
```
